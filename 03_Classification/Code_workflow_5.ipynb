## Search Song

```python
# Ask user and return list of matching songs 

def search_song():
    
    songname = input("Enter your song: ")
    
    playlist_features_list = ["Track","Artist","Album","ID"]
    playlist_df = pd.DataFrame(columns = playlist_features_list)
    
    results = sp.search(q=songname, limit=10)
    songlist = results['tracks']['items']
    
    for track in songlist:
        # Create empty dict
        playlist_features = {}
        # Get metadata
        playlist_features["Track"] = track['name']
        playlist_features["Artist"] = track["album"]["artists"][0]["name"]
        playlist_features["Album"] = track["album"]["name"]
        playlist_features["ID"] = track["id"]
        
        # Concat the dfs
        track_df = pd.DataFrame(playlist_features, index = [0])
        playlist_df = pd.concat([playlist_df, track_df], ignore_index = True)
        
    return playlist_df
```

```python
# Query user for the exact song and return href to the code

def search_song_user():
    df=search_song()
    display(df.drop("ID",axis=1))
    correctsong = int(input("Enter the row number of the song you wanted: "))
    final_choice = df.iloc[correctsong,]
    final_choice_intern = df.iloc[correctsong,3]
    return final_choice_intern
```

## API-Call und Classification

```python
# use href to query spotify for the features
def get_audio_features(spotid):    
    
    #step1

    playlist_features_list = ["track","artist","album", "id","danceability","energy","key","loudness","mode", "speechiness","instrumentalness","liveness","valence","tempo", "duration_ms","time_signature"]
    playlist_df = pd.DataFrame(columns = playlist_features_list)
    
    #step2
    
    result = sp.track(spotid)
    
    # Create empty dict
    result_features = {}
    # Get metadata
    result_features["track"] = str(result['name'])
    result_features["artist"] = result["album"]["artists"][0]["name"]
    result_features["album"] = result["album"]["name"]
    result_features["id"] = result["id"]

    # Get audio features
    audio_features = sp.audio_features(spotid)[0]
    for feature in playlist_features_list[4:]:
        result_features[feature] = audio_features[feature]

    # Concat the dfs
    track_df = pd.DataFrame(result_features, index = [0])
    song_df = pd.concat([playlist_df, track_df], ignore_index = True)

    #Step 3

    return song_df  
```

```python
# load scaler and model to cluster the song

def classification(spotid):
    
    # load function get song data
    song_df = get_audio_features(spotid)
    song_features_df = song_df[['danceability','energy','speechiness','valence','tempo']]
    
    # load pickle loader
    def load(filename = "filename.pickle"): 
        try: 
            with open(filename, "rb") as f: 
                return pickle.load(f) 
        except FileNotFoundError: 
            print("File not found!")
        
    # load scaler and model
    scaler = load("/Users/robertkammerer/Ironhack/06_Week/Labs_Week_6/Spotify_Project/03_Classification/pickle5/scaler.pickle")
    best_model = load("/Users/robertkammerer/Ironhack/06_Week/Labs_Week_6/Spotify_Project/03_Classification/pickle5/kmeans_10.pickle")
    
    # scale features
    song_features_scaled = scaler.transform(song_features_df)
    song_features_scaled_df = pd.DataFrame(song_features_scaled, columns = song_features_df.columns)
    
    # Create clusters
    clusters = best_model.predict(song_features_scaled_df)
    song_df_final = song_df.copy()
    song_df_final["cluster"] = clusters
    
    return list(zip(song_df_final["id"],song_df_final["cluster"]))
    
```

## Recommendation

```python
# Make a song recommendation

def recommendation(id_cluster):
    song_id = id_cluster[0][0]
    cluster = id_cluster[0][1]
    
    if (song_id in database[database['hot/not']=="hot"]['id'].unique()) == True:
        if (cluster in database[database['hot/not']=="hot"]['cluster'].unique()) == True:
            return database[(database['hot/not']=="hot") & (database['cluster']==cluster)].sample()[['track', "artist", "link"]]
        else:
            return database[(database['hot/not']=="not") & (database['cluster']==cluster)].sample()[['track', "artist", "link"]]
    else:
        return database[(database['hot/not']=="not") & (database['cluster']==cluster)].sample()[['track', "artist", "link"]]
```
