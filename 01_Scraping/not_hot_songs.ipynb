```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url90="https://www.morefm.co.nz/home/music/2017/04/the-serious-top-500-of-the-90s-countdown---resurrected.html"
```

```python
html90 = requests.get(url90)
print(html90.status_code)
```

```python
soup90 = BeautifulSoup(html90.text, 'html.parser')
print(soup90.prettify)
```

```python
soup90 = soup90.select("div.c-ArticleBody > p ")
soup90= soup90[4:]
soup90
```

```python
nineties = []

for songs90 in soup90:
    nineties.append(songs90.get_text()[:-7].split(": ")[1].split(" - "))


nineties[489]= 'Jessie - Joshua Kadison'
nineties[489]=nineties[489].split("-")
nineties[481]= 'Ray Of Light - Madonna'
nineties[481]=nineties[481].split(" - ")
nineties[395]= 'C U When U Get There - Coolio'
nineties[395]=nineties[395].split(" - ")
nineties
```

```python
nineties_songs = []
nineties_artist= []

for val in nineties[0:499]:
    nineties_songs.append(val[0])
    nineties_artist.append(val[1])
    
nineties_songs     
```

```python
df90s = pd.DataFrame({"Artist":nineties_artist,"Song":nineties_songs})
display(df90s)
```

```python
url00="https://totototops.wordpress.com/2013/03/29/pitchfork-the-top-500-tracks-of-the-2000s/"
html00 = requests.get(url00)
print(html00.status_code)
soup00 = BeautifulSoup(html00.text, 'html.parser')
print(soup00.prettify)
```

```python
twothou_songs = []
twothou_artists= []

for x in soup00.select("div.entry-content > ol > li")[0:482]:
    twothou_songs.append(x.get_text()[:-7].split(" – ")[1])
    twothou_artists.append(x.get_text()[:-7].split(" – ")[0])


for x in soup00.select("div.entry-content > ol > li")[484:]:
    twothou_songs.append(x.get_text()[:-7].split(" – ")[1])
    twothou_artists.append(x.get_text()[:-7].split(" – ")[0])
    
twothou_songs
twothou_artists
```

```python
df00s = pd.DataFrame({"Artist":twothou_artists,"Song":twothou_songs})
display(df00s)
```
